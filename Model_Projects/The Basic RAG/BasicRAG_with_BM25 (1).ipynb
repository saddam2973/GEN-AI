```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_mistralai import ChatMistralAI

from rank_bm25 import BM25Okapi
import numpy as np
import os
import re


loader = PyPDFLoader("/content/Sri_Harshith_Bollu_Resume (4).pdf")
docs = loader.load()


splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
chunks = splitter.split_documents(docs)
print(f"Loaded {len(docs)} docs, split into {len(chunks)} chunks")
# inspect one
print(chunks[0].page_content[:400])

texts = [c.page_content for c in chunks]
```

```python
# Build BM25 Index
# -------------------------


import re
from rank_bm25 import BM25Okapi

texts = [doc.page_content for doc in chunks]
tokenized_corpus = [re.findall(r"\w+", t.lower()) for t in texts]

bm25 = BM25Okapi(tokenized_corpus)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

print("Creating embeddings... this may take a moment.")
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain_core import vectorstores
# vectorstore = FAISS.from_documents(chunks, embeddings)

chroma_dir = "./chroma_db"

if not os.path.exists(chroma_dir):
  os.makedirs(chroma_dir)

vectorstore = Chroma.from_documents(
      documents=chunks,
      embedding=embeddings,
      persist_directory=chroma_dir
  )
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
```

```python
# -------------------------
# Hybrid Retriever Function
# -------------------------


from collections import OrderedDict

def hybrid_search(query, top_k=3):
    # --- BM25 retrieval ---
    tokenized_query = re.findall(r"\w+", query.lower())
    bm25_scores = bm25.get_scores(tokenized_query)

    bm25_top_indices = np.argsort(bm25_scores)[::-1][:top_k]
    bm25_docs = [chunks[i] for i in bm25_top_indices]

    # --- Vector retrieval (existing retriever) ---
    vector_docs = retriever.invoke(query)


    # --- Fusion (document-level, safe) ---
    fused_docs = OrderedDict()

    for doc in bm25_docs + vector_docs:
        fused_docs[doc.page_content] = doc

    return list(fused_docs.values())[:top_k]
```

```python
# Build Hybrid Retriever Wrapper
# -------------------------
def hybrid_retriever(query):
    results = hybrid_search(query, top_k=3)
    return results
```

```python
from langchain_mistralai import ChatMistralAI

llm = ChatMistralAI(model="mistral-tiny")
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_classic.chains import create_retrieval_chain

prompt = ChatPromptTemplate.from_template("""
Answer the following question based only on the provided context:

<context>
{context}
</context>

Question: {input}
""")
```

```python
document_chain = create_stuff_documents_chain(llm, prompt)
```

```python
# -------------------------
# HYBRID RAG CHAIN
# -------------------------
def hybrid_rag(query):
    docs = hybrid_retriever(query)
    response = document_chain.invoke({"input": query, "context": docs})
    return response
```

```python
rag_chain = create_retrieval_chain(retriever, document_chain)
```

```python
user_query = "What does the document talk about ?"
response = hybrid_rag(user_query)

print("\n--- RESPONSE ---")
print(response)
```
